  Given the `root` of a binary tree, return all root to leaf paths in any order.

  ## Examples

  # The transversal method being used is the depth first search transversal method because each path is explored down to its leaf before tracking.

  Output: [[1, 2, 3], [1, 2, 5], [1, 2, 6], [1, 2, 7]]

  ### Example 2

  ![](./images/q1_ex3.png)

  Input: `root = [10, 9, 7, 8]`

  Output: [[10, 7], [10, 9, 8]]

</details>

#### Starter Code for Question 2

```python
from typing import List

class TreeNode(object):
    def __init__(self, val = 0, left = None, right = None):
        self.val = val
        self.left = left
        self.right = right

def bt_path(root: TreeNode) -> List[List[int]]:
    if not root:
        return []

    paths = []

    def dfs(node, current_path):
        current_path.append(node.val)

        if not node.left and not node.right:
            paths.append(list(current_path))
        else:
            if node.left:
                dfs(node.left, current_path)
            if node.right:
                dfs(node.right, current_path)

        current_path.pop()

    dfs(root, [])
    return paths
```


-   Copy the solution your partner wrote. 


```python
from typing import List #(this is to the cell can run indepedently wihout running the top cell)

def first_duplicate(nums: List[int]) -> int:
    seen = set()
    for num in nums:
        if num in seen:
            return num
        seen.add(num)
        
    return -1
```


-   Explain why their solution works in your own words.


```python
# This function tracks numbers we’ve seen using a set (which has fast lookups). As we move through the list:
# If a number is already in the set, it means we’ve seen it before—so we return it immediately.
# If it’s new, we add it to the set and continue.
# This approach guarantees we find the first repeated number based on the order we traverse the list.
```


-   Explain the problem’s time and space complexity in your own words.


```python
# Time Complexity: O(n) Because we loop through the list once and set lookups are constant time.
# Space Complexity: O(n) In the worst case (no duplicates), we store every number in the set.
# This is much more efficient than the brute force method, which would require checking every pair of values (O(n²)).
```
